```python
df.dropna(inplace=True)
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype('int')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')
```

```python
df.rename(columns={'Qty':'Quantity'},inplace=True)
```

```python
df.columns
```

```python
df.describe()
```

```python
order = ['XS','S','M','L','XL','XXL','3XL','4XL','5XL','6XL','Free']
ax = sns.countplot(x='Size',data = df,order=order)
for bars in ax.containers:
    ax.bar_label(bars)
```

### Note : From above graph we can see that most buyers buy M size

```python
size_quantity = df.groupby('Size')['Quantity'].sum().sort_values(ascending=False)
size_quantity_df = size_quantity.reset_index()
ax=sns.barplot(x='Size',y='Quantity',data=size_quantity_df)
for container in ax.containers:
    ax.bar_label(container)
```

### Note : From above graph we can see that the M size has the highest number of quantity ordered by the buyers

```python
plt.figure(figsize=(20, 6))
ax = sns.countplot(x='Courier Status', hue='Status', data=df)
for container in ax.containers:
    ax.bar_label(container)
```

```python
ax = sns.countplot(x='Category',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

### Note : From above graph we can see that most of the buyers buy T-shirt

```python
category_amount = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
category_amount_df = category_amount.reset_index()
plt.figure(figsize=(50, 20))
ax=sns.barplot(x='Category',y='Amount',data=category_amount_df)
ax.set_xlabel('Category', fontsize=40)
ax.set_ylabel('Total Amount', fontsize=40)
ax.tick_params(axis='x', labelsize=40)
ax.tick_params(axis='y', labelsize=40)
for index, value in enumerate(category_amount_df['Amount']):
    plt.text(index, value + 1000, f'{int(value):,}', ha='center', va='bottom',fontsize=30)
```

### Note : From above graph we can see that highest total  amount is generated by T-shirts

```python
x = df['Category']
y = df['Size']
plt.scatter(x,y)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
```

### Note: From above Scatter Plot we can see what Sizes are available in which Category

```python
b2b = df['B2B'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(b2b, labels=b2b.index, autopct='%1.1f%%', startangle=40, textprops={'fontsize': 14})
plt.title('B2B Distribution', fontsize=18)
plt.show()
```

### Note: From above pie chart we can see that 99.2% buyers are retailers and remaining 0.8% are B2B

```python
top_10_states = df['ship-state'].value_counts().nlargest(10).index
filtered_df = df[df['ship-state'].isin(top_10_states)]
plt.figure(figsize=(30, 8))
ax = sns.countplot(x='ship-state', data=filtered_df, order=top_10_states)
ax.set_xlabel('States', fontsize=15)
ax.set_ylabel('Count', fontsize=15)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
for bars in ax.containers:
    ax.bar_label(bars,fontsize=15)
```

### Note : From above graph we can see that Maharashtra state has most number of buyers

```python
state_amount = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)
state_amount_df = state_amount.reset_index()
plt.figure(figsize=(20, 6))
ax=sns.barplot(x='ship-state',y='Amount',data=state_amount_df)
ax.set_xlabel('States', fontsize=10)
ax.set_ylabel('Total Amount', fontsize=10)
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
for index, value in enumerate(state_amount_df['Amount']):
    plt.text(index, value + 1000, f'{int(value):,}', ha='center', va='bottom')
```

### Note : From above graph we can see that the Maharashtra state has the highest Total Amount generated

```python
top_10_city = df['ship-city'].value_counts().nlargest(10).index
filtered_df = df[df['ship-city'].isin(top_10_city)]
plt.figure(figsize=(30, 8))
ax = sns.countplot(x='ship-city', data=filtered_df, order=top_10_city)
ax.set_xlabel('Cities', fontsize=15)
ax.set_ylabel('Count', fontsize=15)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
for bars in ax.containers:
    ax.bar_label(bars,fontsize=15)
```

### Note : From above graph we can see that the Bengaluru city has the most number of buyers
